```python
import numpy as np
lst_marks=[45,32,56,75,54,32,89,67,54,45,98,99,67,74]
min,Q1,medium,Q3,max = np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
min,Q1,medium,Q3,max 
```

```python
IQR = Q3-Q1
```

```python
print(IQR)
```

```python
lower_fence = Q1-1.5*(IQR)
higher_fence = Q1+1.5*(IQR)
```

```python
print(lower_fence)
print(higher_fence)
```
