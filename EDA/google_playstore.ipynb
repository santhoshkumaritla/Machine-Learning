```python
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']
```

```python
print('We have {} numerical Features:{}'.format(len(numeric_features),numeric_features))
print('We have {} Catergorivcal Features:{}'.format(len(categorical_features),categorical_features))
```

```python
##proportion of count on categorical columns
for col in categorical_features:
  print(df[col].value_counts(normalize=True)*100)
  print("------------------------------------")
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate analysis of numerial Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i + 1)
    sns.kdeplot(x=df_copy[numeric_features[i]], fill=True, color='r')  # Fixed args
    plt.xlabel(numeric_features[i])

plt.tight_layout()
```

```python
plt.figure(figsize=(20,15))
plt.suptitle('Univariate analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = ['Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i + 1)
    sns.countplot(x=df_copy[category[i]], palette='Set2')
    plt.xticks(rotation=45)

plt.tight_layout()
```
